```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_boston
bostan_df=load_boston()
print(bostan_df.keys())
```

```python
print(bostan_df["DESCR"])
```

```python
data=pd.DataFrame(data=bostan_df.data,columns=bostan_df.feature_names)
print(data.head())
```

```python
data["Price"]=bostan_df.target
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
print(data.isnull().count())
```

```python
data.corr()
```

```python
import seaborn as sns
sns.pairplot(data.corr())
```

```python
plt.subplot(1,2,1)
plt.scatter(data["CRIM"],data["Price"],c='r')
plt.xlabel("CRIM")
plt.ylabel("Price(in 1000$)")

plt.subplot(1,2,2)
plt.scatter(data["B"],data["Price"])
plt.xlabel("B")
plt.ylabel("Price(in 1000$)")
plt.show()
```

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
print(X_train)
```

```python
import pickle
scaler=pickle.dump(sc,open("scaler.pkl","wb"))
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

```python
y_pred=regressor.predict(X_test)
print(y_pred)
```

```python
plt.scatter(y_pred,y_test)
plt.show()
```

```python
residual=y_pred-y_test
```

```python
print(residual)
```

```python
plt.scatter(y_pred,residual)
plt.show()
```

```python
sns.displot(residual,kind="kde")

plt.show()
```

```python
import pickle
pickle.dump(regressor,open("regmodel.pkl","wb"))
pkld_model=pickle.load(open("regmodel.pkl","rb"))
pkld_model.predict([X_test[0]])
```
